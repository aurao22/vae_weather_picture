### Charger les images dans une DF

```python
# récupération données
img_extensions = [".jpg", ".jpeg", ".png"]
images = []

for ext in img_extensions:
    images.extend(vae_w.get_dir_files(source_data_path, endwith=ext))

print(len(images))
images
```

```python
def get_picture_dic(images, load_bw=True, nb_desc=300, verbose=0):
    img_dic = defaultdict(list)

    pattern = vae_w.get_regex_alphabetique_simple()

    for img_path in images:
        img_dic["file_name"].append(img_path)
        name = img_path.lower().split(".")[0]
        name = re.sub(pattern, '', name).strip()
        img_dic["y"].append(name)
        
        im = None
        if load_bw:
            im = cv2.imread(source_data_path + img_path, 0)
        else:
            im = cv2.imread(source_data_path + img_path)
        
        img_dic["width"].append(im.shape[0])
        img_dic["higth"].append(im.shape[1])
        
        img_dic["format"].append(img_path.upper().split(".")[1])
        # img_dic["mode"].append(im.mode)
        # img_dic["Ratio"].append(round(im.shape[1]/im.shape[0], 2))
        # TODO récupérer les descripteurs
        # Equalise des images :
        equ = cv2.equalizeHist(im)
        sift = cv2.SIFT_create(nb_desc)
        kp, des = sift.detectAndCompute(equ,None)

        i = 0
        for pixel in des:
            i += 1
            img_dic["DESC_"+str(i)].append(pixel)

    return img_dic
```

```python
img_dic = get_picture_dic(images,load_bw=True, nb_desc=300, verbose=verbose)
df_origin = pd.DataFrame.from_dict(img_dic)
print(df_origin.shape)
df_origin.columns
```
